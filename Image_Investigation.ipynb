```python
file_names = []
severity = []
lesions = []
file_sev = []
for i in range(4):
  with open(f'/content/Classification/NNEW_trainval_{i}.txt') as f:
      for line in f.readlines():
        name, sev, lesion = line.split()
        file_names.append(f'/content/Classification/JPEGImages/{name}')
        severity.append(sev)
        lesions.append(lesion)
        file_sev.append((name,sev))
```

```python
sev_count = {}
for i in range(len(severity)):
  if severity[i] not in sev_count.keys():
    sev_count[severity[i]] = 0
    sev_count[severity[i]] += 1
  else:
    sev_count[severity[i]] += 1
```

```python
print(sev_count)
plt.bar(sev_count.keys(),sev_count.values())
plt.title('Label Distribution')
plt.xlabel('Severity')
plt.ylabel('Count')
```

```python
import os
import pandas as pd
from torchvision.io import read_image
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torchvision import transforms

class AcneDataset(Dataset):
    def __init__(self, labels, image_paths, transform=None, target_transform=None):
        self.img_labels = labels
        self.img_dir = image_paths
        self.transform = transform
        self.target_transform = target_transform

    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = self.img_dir[idx]
        image = read_image(img_path)
        label = torch.Tensor([int(self.img_labels[idx])])
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        
        return image, label
```

```python
train_set = AcneDataset(severity,file_names,transform=transforms.Compose([transforms.Resize((256, 256))]))
train_dataloader = DataLoader(train_set,batch_size=64)
```

```python
train_features, train_labels = next(iter(train_dataloader))
print(f"Feature batch shape: {train_features.size()}")
print(f"Labels batch shape: {train_labels.size()}")
img = train_features[0].squeeze()
label = train_labels[0]
plt.imshow(img.T, cmap="gray")
plt.show()
print(f"Label: {label}")
```
